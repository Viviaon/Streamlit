```python
import streamlit as st
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
link = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/cars.csv"
df_car = pd.read_csv(link)
```

```python
from IPython.display import display
# Used to display basic information and first lines of a df

display(df_car.info(),
        df_car.head(5))
```

```python
mat = px.scatter_matrix(df_car)

mat.update_layout(
    title = 'Correlation Matrix',
    width = 1200,
    height = 1200,
)

mat.update_traces(diagonal_visible = False)
mat.show()
```

```python
fig, ax = plt.subplots(1, 2)

fig1 = plt.subplot(121)
fig1 = px.scatter(df_car, x = "hp", y = "mpg", 
                  color = "continent", 
                  color_discrete_map = {"US.": "red",
                                        "Europe." : "blue",
                                        "Japan.": "green"})

fig2 = plt.subplot(122)
fig2 = px.histogram(df_car, x = "cylinders", color = "continent")

fig1.show()
fig2.show()
```
